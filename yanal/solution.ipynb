**Problem A: Key Word in Context (KWIC)**

***Method 3: Pipes-and-Filters***

```python
# Document Processor
def process_document(text):
    # Process the document (e.g., remove punctuation) and return the text
    return text

# Shifter
def perform_shift(document):
    lines = document.split('\n')
    shifted_lines = []

    for line in lines:
        words = line.split()
        for i in range(len(words)):
            shifted_line = ' '.join(words[i:] + words[:i])
            shifted_lines.append(shifted_line)

    return '\n'.join(shifted_lines)

# Alphabetizer
def sort_lines(document):
    lines = document.split('\n')
    sorted_lines = sorted(lines)
    return '\n'.join(sorted_lines)

# Usage:
text = "This is a sample document with keywords.\nPlease process this document."
processed_text = process_document(text)
shifted_text = perform_shift(processed_text)
sorted_text = sort_lines(shifted_text)
print(sorted_text)
```

**Problem B: Eight Queens (8Q)**


***Method 2: Main/Subroutine with Stepwise Refinement***

```python
class Chessboard:
    def __init__(self, size):
        self.size = size
        self.board = [[0] * size for _ in range(size)]
        self.solutions = []

    def is_safe(self, row, col):
        for i in range(col):
            if self.board[row][i] == 1:
                return False
            if row - i >= 0 and self.board[row - i][col - i] == 1:
                return False
            if row + i < self.size and self.board[row + i][col - i] == 1:
                return False
        return True

    def solve(self, col):
        if col == self.size:
            solution = []
            for row in range(self.size):
                solution.append(''.join(['Q' if self.board[row][i] == 1 else '.' for i in range(self.size)]))
            self.solutions.append(solution)
            return
        for row in range(self.size):
            if self.is_safe(row, col):
                self.board[row][col] = 1
                self.solve(col + 1)
                self.board[row][col] = 0

def eight_queens_solver(size):
    chessboard = Chessboard(size)
    chessboard.solve(0)
    return chessboard.solutions

# Usage:
board_size = 8
solutions = eight_queens_solver(board_size)
for solution in solutions:
    for row in solution:
        print(row)
    print()
```
